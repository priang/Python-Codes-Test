```python
from datetime import date
```

```python
data_hoje = date.today()
dia_hoje = data_hoje.day
mes_hoje = data_hoje.month
ano_hoje = data_hoje.year

print(dia_hoje,mes_hoje,ano_hoje)
```

```python
data_inicio = '20/01/1999'
dia_ini = int(data_inicio[:2])
mes_ini = int(data_inicio[3:5])
ano_ini = int(data_inicio[6:])

print(dia_ini,mes_ini,ano_ini)
```

```python
dias_mes = [31,28,31,30,31,30,31,31,30,31,30,31]
```

Calculando o número de dias do começo do ano até hoje:

```python
dias_passados = sum(dias_mes[:mes_hoje-1])

if ano_hoje%4==0 and (ano_hoje%100!=0 or ano_hoje%400==0) and mes_hoje > 2:
  dias_passados = dias_passados+1

dias_passados = dias_passados + dia_hoje

dias_passados
```

Calculando o número de dias até o fim do ano da data inicial (data de nascimento)

```python
dias_faltantes = sum(dias_mes[mes_ini-1:])

if ano_ini%4==0 and (ano_ini%100!=0 or ano_ini%400==0) and mes_ini <= 2:
 dias_faltantes = dias_faltantes + 1

dias_faltantes = dias_faltantes - dia_ini

dias_faltantes
```

Calculando o número de dias dos anos completos:

```python
total = 0
for ano in range(ano_ini+1,ano_hoje):
  total = total + 365
  if ano%4==0 and (ano%100!=0 or ano%400==0):
     total = total + 1
total
```

```python
total = total + dias_passados + dias_faltantes
total
```
